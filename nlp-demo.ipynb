```python
# Install tensorflow package
# %pip install --upgrade tensorflow


import json
import pandas as pd
import re
import numpy as np
import string
from tensorflow.keras.models import load_model 
```

```python

# model1
model1 = load_model(r"model1/model.keras")
# model2
model2 = load_model(r"model2/model.keras")


# load data of model1
with open(r"model1/tag2idx_model1.json", "r") as file:
    tag2idx1 = json.load(file)
with open(r"model1/word2idx_model1.json", "r") as file:
    word2idx1 = json.load(file)


# load data of model2
with open(r"model2/tag2idx_model2.json", "r") as file:
    tag2idx2 = json.load(file)
with open(r"model2/word2idx_model2.json", "r") as file:
    word2idx2 = json.load(file)

print(len(word2idx1))
print(len(word2idx2))
```

```python
# load dev data
dev_data = pd.read_json("dataset\PIZZA_dev.json", lines=True)
src_dev=dev_data["dev.SRC"].to_numpy()
top_dev=dev_data["dev.TOP"]
MAX_LEN=52
words = word2idx1.keys()
with open("top_dev.txt", "w") as file:
    for top in top_dev:
        file.write(top + "\n")
```

```python
idx2word1 = {v: k for k, v in word2idx1.items()}  
idx2word2 = {v: k for k, v in word2idx2.items()}  
idx2tag = {v: k for k, v in tag2idx1.items()}  
idx2tag2 = {v: k for k, v in tag2idx2.items()}  
```

```python
def convert_to_json(new_predicted1, new_predicted2, model1_words, unk):
    # Helper variables
    current_group = None
    current_entity = {}
    all_toppings = []
    all_styles = []
    json_result = {"ORDER": {"PIZZAORDER": [], "DRINKORDER": []}}
    
    try:
        # Iterate through labels and words
        for i, (entity_label, group_label) in enumerate(zip(new_predicted1, new_predicted2)):
            word = model1_words[i]
            if word == "<UNK>":
                word = unk.pop(0)
            
            if group_label.startswith("B-"):
                # Save the previous group entity if it exists
                if current_group == "PIZZAORDER" :
                    if all_toppings:
                        current_entity["ALLTOPPING"] = all_toppings
                        all_toppings = []
                    if all_styles:
                        current_entity["STYLE"] = all_styles
                        all_styles = []
                    json_result["ORDER"]["PIZZAORDER"].append(current_entity)
                elif current_group == "DRINKORDER" and current_entity:
                    json_result["ORDER"]["DRINKORDER"].append(current_entity)
                
                # Start a new entity
                current_group = group_label[2:]
                current_entity = {}
        
            if entity_label == "B-NUMBER" and  current_entity.get("NUMBER") is None:
                current_entity["NUMBER"] = word
            elif entity_label == "B-SIZE":
                current_entity["SIZE"] = word
            
            elif entity_label in ("B-STYLE"):
                all_styles.append({"NOT": False, "TYPE": word})
            elif entity_label in ("I-STYLE"):
                all_styles[-1]["TYPE"] += " " + word
            elif entity_label in ("B-NOT-STYLE"):
                all_styles.append({"NOT": True, "TYPE": word})
            elif entity_label in ("I-NOT-STYLE"):
                all_styles[-1]["TYPE"] += " " + word
    
            elif entity_label == "B-QUANTITY":
                all_toppings.append({"NOT": False, "QUANTITY": word, "TOPPING": None})
            elif entity_label == "I-QUANTITY":
                all_toppings[-1]["QUANTITY"] += " " + word
            elif entity_label == "B-TOPPING" and all_toppings and all_toppings[-1]["TOPPING"] is None:
                all_toppings[-1]["TOPPING"] = word
            elif entity_label == "B-TOPPING":
                all_toppings.append({"NOT": False, "QUANTITY": None, "TOPPING": word})
            elif entity_label == "I-TOPPING":
                all_toppings[-1]["TOPPING"] += " " + word
            elif entity_label == "B-NOT-TOPPING" and all_toppings and all_toppings[-1]["TOPPING"] is None:
                all_toppings[-1]["NOT"] = True
                all_toppings[-1]["TOPPING"] = word
            elif entity_label == "B-NOT-TOPPING":
                all_toppings.append({"NOT": True, "QUANTITY": None, "TOPPING": word})
            elif entity_label == "I-NOT-TOPPING":
                all_toppings[-1]["TOPPING"] += " " + word
            elif entity_label == "B-CONTAINERTYPE":
                current_entity["CONTAINERTYPE"] = word
            elif entity_label in ("B-DRINKTYPE", "I-DRINKTYPE"):
                current_entity["DRINKTYPE"] = current_entity.get("DRINKTYPE", "") + " " + word
        
        # Add the last group entity
        if current_group == "PIZZAORDER" :
            if all_toppings:
                current_entity["ALLTOPPING"] = all_toppings
            if all_styles:
                current_entity["STYLE"] = all_styles
            json_result["ORDER"]["PIZZAORDER"].append(current_entity)
        elif current_group == "DRINKORDER" and current_entity:
            json_result["ORDER"]["DRINKORDER"].append(current_entity)
    
        # Remove empty orders
        if 'PIZZAORDER' in json_result['ORDER'] and not json_result['ORDER']['PIZZAORDER']:
            del json_result['ORDER']['PIZZAORDER']
        if 'DRINKORDER' in json_result['ORDER'] and not json_result['ORDER']['DRINKORDER']:
            del json_result['ORDER']['DRINKORDER']

    
    except Exception as e:
        return {}
    
    return json_result
```

```python
def clean_alltopping(json_data):
    for pizza in json_data.get("ORDER", {}).get("PIZZAORDER", []):
        if "ALLTOPPING" in pizza:
            # Filter out elements with a QUANTITY but no TOPPING
            pizza["ALLTOPPING"] = [
                topping for topping in pizza["ALLTOPPING"]
                if not (topping.get("QUANTITY") and not topping.get("TOPPING"))
            ]
    return json_data
```

```python
def predict(test_src):
   
    
    for src in test_src:
        allowed_punctuation = "-_'"
        translation_table = src.maketrans('', '', string.punctuation.replace(allowed_punctuation, ''))
        src=src.translate(translation_table)
        unk=[]
        src_words_orginal = src.split() 
        if(len(src_words_orginal)>MAX_LEN):
            continue
        # replace unknown words
        for word in src_words_orginal:
            if word not in words:
                str=f"(?:(?<=\\s)|^){word}(?:(?=\\s)|$)"
                src=re.sub(str,"<UNK>",src,flags=re.MULTILINE,count=1)
                unk.append(word)
        src_words=src.split()
        # get the words indecies
        src_indecies = [word2idx1.get(w) for w in src_words]
        # pad
        pad_token = word2idx1.get("<pad>")  
        src_indecies = src_indecies + [pad_token] * (MAX_LEN - len(src_words))
        src_indecies = np.expand_dims(src_indecies, axis=0)  
        # prediction
        p = model1.predict(src_indecies)  
        p = np.argmax(p, axis=-1)  # Get the predicted tag indices
        
        predicted1=[]
        
        for i,tag_idx in enumerate(p[0]):
            tag_begin=idx2tag[tag_idx]
            if tag_begin[0:2]=="I-" and i>0 and  tag_begin[2:-1]!=predicted1[-1][2:-1] :
                tag_begin=re.sub("I-","B-",tag_begin)
             
            predicted1.append(tag_begin) 


        
        ##################################################
        
        new_src_indecies=[]
        for w,t in zip(src_indecies[0],predicted1):
            if t !="O":
                new_src_indecies.append(w)

        model1_words=[]
        for w in new_src_indecies:
            model1_words.append(idx2word1[w])

    
        model2_src_indecies=[]
        for w in model1_words:
            model2_src_indecies.append(word2idx2.get(w,word2idx2["<UNK>"]))

       
        ###################################################
        pad_token=word2idx2.get("<pad>")  
        model2_src_indecies = model2_src_indecies + [pad_token] * (MAX_LEN - len(model2_src_indecies))
        model2_src_indecies = np.expand_dims(model2_src_indecies, axis=0)  # Shape: (1, 25)
        p = model2.predict(model2_src_indecies)  
        p = np.argmax(p, axis=-1)  # Get the predicted tag indices
        predicted2=[]
        for i,w in enumerate(p[0]):  # p is likely a batch, so take the first sequence
            tag_begin=idx2tag2[w]
            if tag_begin[0:2]=="I-" and i>0 and tag_begin[2:-1]!=predicted2[-1][2:-1] :
                tag_begin=re.sub("I-","B-",tag_begin)
            if tag_begin=="B-QUANTITY"  and i>0 and predicted2[-1]== "I-QUANTITY":
                tag_begin=re.sub("B-","I-",tag_begin)
            predicted2.append(tag_begin) 

        
        new_predicted1=[]
        new_predicted2=[]
        for p1 in predicted1:
            if p1!="O":
                 new_predicted1.append(p1)
        for p2 in predicted2:
            if p2!="O":
                 new_predicted2.append(p2)
                
        # conver to json
        json_result=convert_to_json(new_predicted1,new_predicted2,model1_words,unk)
        json_result=clean_alltopping(json_result)
        print(json.dumps(json_result, indent=4))
        


        
       
        
                
                
```
